```python
import pandas as pd
```

```python
bar_df = pd.read_csv('data_sources/Atx_bar.csv')
club_df = pd.read_csv('data_sources/Atx_club.csv')
rest_df = pd.read_csv('data_sources/Atx_restaurant.csv')
bar_df.head()
```

```python
club_df.head()
```

```python
#Get Rid of Useless Columns

del bar_df['Unnamed: 0']
#del bar_df['barlatitude']
#del bar_df['barlongitude']
del bar_df['barName']
del club_df['Unnamed: 0']
#del club_df['Club latitude']
#del club_df['Club longitude']
del club_df['Club Name']
del rest_df['Unnamed: 0']
#del rest_df['Restaurant_latitude']
#del rest_df['Restaurant_longitude']
del rest_df['Restaurant_Name']
bar_df.head()
```

```python
#Group By Count and Average by ZipCode for totals and avg rating

bar_count = bar_df.groupby(['Zip code']).count()
bar_avg = bar_df.groupby(['Zip code']).mean()
club_count = club_df.groupby(['Zip code']).count()
club_avg = club_df.groupby(['Zip code']).mean()
rest_count = rest_df.groupby(['Zip code']).count()
rest_avg = rest_df.groupby(['Zip code']).mean()

bar_avg.head()
```

```python
#Merge Like DFs

rest_merge = pd.merge(rest_count, rest_avg, on= 'Zip code')
bar_merge = pd.merge(bar_count, bar_avg, on= 'Zip code')
club_merge = pd.merge(club_count, club_avg, on= 'Zip code')
```

```python
#Merge All DFs

rest_bar = pd.merge(rest_merge, bar_merge, on = 'Zip code', how = 'outer')
nightlife_df = pd.merge(rest_bar, club_merge, on = 'Zip code', how = 'outer')
nightlife_df
```

```python
#Reset Index
#nightlife_df = nightlife_df.reset_index()

#Delete Useless Columns

#del nightlife_df['Restaurant_latitude_x']
#del nightlife_df['Restaurant_longitude_x']
#del nightlife_df['barlatitude_x']
#del nightlife_df['barlongitude_x']
#del nightlife_df['Club latitude_x']
#del nightlife_df['Club longitude_x']


#Rename all the columns

nightlife_df.columns = ['Zip Code','Restaurant Count', 'Restaurant Rating', 'Restaurant Latitude',
                        'Restaurant Longitude', 'Bar Count', 'Bar Rating', 'Bar Latitude', 'Bar Longitude',
                        'Club Count', 'Club Rating', 'Club Latitude', 'Club Longitude']
nightlife_df.head()
```

```python
#Fill Na Values with Zeroes

nightlife_df = nightlife_df.fillna(0)

nightlife_df['Latitude'] = nightlife_df.loc[nightlife_df['Club Latitude'] != 0]
nightlife_df['Latitude'] = nightlife_df.loc[nightlife_df['Bar Latitude'] != 0]
nightlife_df['Latitude'] = nightlife_df.loc[nightlife_df['Restaurant Latitude'] != 0]

nightlife_df.head()
```

```python
#Add Bar and Restaurant Combined

nightlife_df['Bar/Restaurant Count'] = (nightlife_df['Restaurant Count'] + nightlife_df['Bar Count']
                                         + nightlife_df['Club Count'])

#Just Bars and Restaurants

nightlife_df['Bar/Restaurant Rating'] = (((nightlife_df['Restaurant Rating'] * nightlife_df['Restaurant Count'])
                                   +(nightlife_df['Bar Rating'] * nightlife_df['Bar Count']))/nightlife_df['Bar/Restaurant Count']) * 2

#Add Total Count Column

nightlife_df['Total Nightlife Count'] = (nightlife_df['Restaurant Count'] + nightlife_df['Bar Count']
                                         + nightlife_df['Club Count'])
```

```python
#Add Total Rating Column

nightlife_df['Combined Rating'] = (((nightlife_df['Restaurant Rating'] * nightlife_df['Restaurant Count'])
                                   +(nightlife_df['Bar Rating'] * nightlife_df['Bar Count']) +
    (nightlife_df['Club Rating'] * nightlife_df['Club Count']))/nightlife_df['Total Nightlife Count']) * 2

```
