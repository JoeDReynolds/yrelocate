```python
import pandas as pd
```

```python
zip_codes = pd.read_csv('data_sources/aus_zip_codes.csv')
bar_df = pd.read_csv('data_sources/bar.csv')
club_df = pd.read_csv('data_sources/club.csv')
zip_codes.head()
```

```python
zip_code = zip_codes.loc[zip_codes["ZipCodeType"] == "STANDARD", :]
zip_code_empty = zip_code.loc[zip_codes['Decommisioned'] == False, :]
```

```python
del zip_code_empty['Location']
del zip_code_empty['Decommisioned']
del zip_code_empty['ZipCodeType']
del zip_code_empty['RecordNumber']
del zip_code_empty['Unnamed: 0']
zip_code_empty.head()
```

```python
zip_code_empty.count()
```

```python
del bar_df['latitude']
del bar_df['longitude']
del bar_df['Unnamed: 0']
bar_df.head()
```

```python
group_bars = bar_df.groupby(['Zip code']).count()
```

```python
group_bars.head()
```

```python
del club_df['latitude']
del club_df['longitude']
del club_df['Unnamed: 0']
club_df.head()
```

```python
group_clubs = club_df.groupby(['Zip code']).count()
group_clubs.head()
```

```python
nightlife_merge = pd.merge(group_clubs, group_bars, on = 'Zip code', how = "outer")
```

```python
nightlife_merge.head()
```

```python
nightlife_merge['Nightlife Count'] = nightlife_merge['Name_x'] + nightlife_merge['Name_y']
```

```python
night = nightlife_merge.reset_index()

del night['Name_x']
del night['Name_y']
night.columns = ['Zipcode', 'Nightlife Count']
night = night[pd.to_numeric(night['Zipcode'], errors='coerce').notnull()]
night['Zipcode'] = night['Zipcode'].astype(int)
```

```python
zip_code_nightlife = pd.merge(zip_code_empty, night, on = 'Zipcode', how = 'left')
zip_code_nightlife.count()
```

```python
rest_df = pd.read_csv('data_sources/restaurant.csv')
del rest_df['latitude']
del rest_df['longitude']
del rest_df['Unnamed: 0']
rest_df.count()
```

```python
group_rest = rest_df.groupby(['Zip code']).count()
rest = group_rest.reset_index()
rest.columns= ['Zipcode', 'Name']
rest.head()
#zip_code_nightrest = pd.merge(zip_code_nightlife, rest, on = "Zipcode")
```

```python
zip_code_nightrest = pd.merge(zip_code_nightlife, rest, on = 'Zipcode', how = 'left')
zip_code_nightrest = zip_code_nightrest.rename(columns = {'Name' : 'Restaurant Count'})
zip_code_nightrest
```
