```python
#Import Dependencies
import pandas as pd
```

```python
zip_codes = pd.read_csv('data_sources/free-zipcode-database.csv')
zip_codes.head()
```

```python
del zip_codes['LocationType']
del zip_codes['Xaxis']
del zip_codes['WorldRegion']
del zip_codes['LocationText']
del zip_codes['TaxReturnsFiled']
del zip_codes['TotalWages']
del zip_codes['Notes']
del zip_codes['Yaxis']
del zip_codes['Zaxis']

zip_codes.head()
```

```python
tx_zip_codes = zip_codes.loc[zip_codes['State'] == 'TX', :]
tx_zip_codes["City"].value_counts()
```

```python
aus_zip_codes = tx_zip_codes.loc[tx_zip_codes['City'] == "AUSTIN" , : ]
aus_zip_codes.head()
```

```python
#Create CSV Files From DataFrames

aus_zip_codes.to_csv("data_sources/aus_zip_codes.csv")
tx_zip_codes.to_csv("data_sources/tx_zip_codes.csv")
zip_codes.to_csv("data_sources/clean_zip_codes.csv")
```
